```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("/Users/ja/Deep-Learning-Projects/lena10.jpg") 
```

```python
plt.imshow(img)
```

```python
img2 = cv2.copyMakeBorder(img,0,0,100,100,cv2.BORDER_CONSTANT,value=[0,0,0])
```

```python
plt.imshow(img2)
```

```python
img3 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img3)
```

```python
img4 = img3 * 1.5 + 50
```

```python
plt.imshow(img4)
```

```python
blur = cv2.GaussianBlur(img4, (7,5), 1)
```

```python
plt.imshow(blur)
```

```python
img5 = cv2.resize(blur,None,fx=2,fy=2,interpolation=cv2.INTER_CUBIC)
```

```python
plt.imshow(img5)
```

```python
h,w = img5.shape
M = cv2.getRotationMatrix2D((w/2,h/2),180,1)
img6 = cv2.warpAffine(img5,M,(w, h))
```

```python
plt.imshow(img6)
```
